### **Задача 1**  
**Условие:**  
Предполагается увеличение доли фирмы на рынке с 10 до 12 % при емкости рынка 175 тыс. штук продукта. Рассчитать дополнительную прибыль фирмы в предстоящем году, если прибыль на одно изделие составляет 98 руб., а емкость рынка не изменяется. Затраты на маркетинговые усилия в расчете на год составляют 182 тыс. руб.  

**Решение:**  
1. Определяем объем продаж до и после роста доли рынка:  
   - До: 175000 × 0.10 = 17500 штук  
   - После: 175000 × 0.12 = 21000 штук  
2. Дополнительные продажи:  
   21000 - 17500 = 3500 штук  
3. Дополнительная прибыль без учета затрат:  
   3500 × 98 = 343000 руб.  
4. Чистая дополнительная прибыль:  
   343000 - 182000 = 161000 руб.  

**Вывод:** Дополнительная прибыль фирмы составит **161 000 руб.**

### **Задача 2**  
**Условие:**  
По мере роста конкуренции доля компании ежегодно снижается на 5 %, но рынок увеличивается на 200 тыс. ед. в год.  
- Текущая емкость рынка: 500 тыс. ед.  
- Текущая доля компании: 35 %  
- Прибыль с единицы: 2 долл.  

**Решение:**  

| Год | Доля рынка, % | Объем продаж, ед. | Валовая прибыль, тыс. долл. |
|----|-------------|--------------|--------------------|
| 1  | 35         | 500000 × 0.35 = 175000 | 175000 × 2 = 350  |
| 2  | 30         | 700000 × 0.30 = 210000 | 210000 × 2 = 420  |
| 3  | 25         | 900000 × 0.25 = 225000 | 225000 × 2 = 450  |
| 4  | 20         | 1100000 × 0.20 = 220000 | 220000 × 2 = 440  |
| 5  | 15         | 1300000 × 0.15 = 195000 | 195000 × 2 = 390  |

**Вывод:** Через 5 лет валовая прибыль снизится до **390 тыс. долл.**

### **Задача 3**  
**Условие:**  
Фирма работает в 3 сегментах. Определить объем продаж.  

**Решение:**  
1. **Первый сегмент:**  
   - Новый объем рынка: 1980 × 1.03 = 2039.4 тыс. шт.  
   - Новая доля фирмы: 54 × 1.05 = 56.7 тыс. шт.  

2. **Второй сегмент:**  
   - Новый объем рынка: 75 / 0.04 = 1875 тыс. шт.  
   - Новый объем рынка после роста: 1875 × 1.11 = 2081.25 тыс. шт.  
   - Новый объем продаж: 2081.25 × 0.08 = 166.5 тыс. шт.  

3. **Третий сегмент:**  
   - Новый объем продаж: 45 × 0.12 - 1 = 4.4 тыс. шт.  

**Общий объем продаж:**  
56.7 + 166.5 + 4.4 = 227.6 тыс. шт.  

**Вывод:** Новый объем продаж фирмы составит **227.6 тыс. шт.**

### **Задача 4**  
**Условие:**  
Цена = 90 ДЕ, предельные издержки = 60 ДЕ.  

**Формула индекса Лернера:**  

L = (P - MC) / P
