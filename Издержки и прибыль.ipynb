```python
#  
price_per_kg = 2000  #    
waste_price = 800    #    
total_waste = 3000    #     

#   
orders = {
    "1": {"output": 3000, "norm_rate": 4, "actual_rate": 4},
    "2": {"output": 2000, "norm_rate": 20, "actual_rate": 17},
    "3": {"output": 1000, "norm_rate": 2, "actual_rate": 3}
}

# 1.   
for order in orders.values():
    order["norm_kg"] = order["output"] * order["norm_rate"]
    order["norm_rub"] = order["norm_kg"] * price_per_kg

# 2.   
for order in orders.values():
    order["actual_kg"] = order["output"] * order["actual_rate"]
    order["actual_rub"] = order["actual_kg"] * price_per_kg

# 3.       
total_norm_kg = sum(o["norm_kg"] for o in orders.values())

# 4.  
for order in orders.values():
    order["waste_kg"] = (order["norm_kg"] / total_norm_kg) * total_waste
    order["waste_rub"] = order["waste_kg"] * waste_price

# 5.   
for order in orders.values():
    order["total_kg"] = order["actual_kg"] - order["waste_kg"]
    order["total_rub"] = order["actual_rub"] - order["waste_rub"]

# 6.   
totals = {
    "output": sum(o["output"] for o in orders.values()),
    "norm_kg": sum(o["norm_kg"] for o in orders.values()),
    "norm_rub": sum(o["norm_rub"] for o in orders.values()),
    "actual_kg": sum(o["actual_kg"] for o in orders.values()),
    "actual_rub": sum(o["actual_rub"] for o in orders.values()),
    "waste_kg": total_waste,
    "waste_rub": total_waste * waste_price,
    "total_kg": sum(o["total_kg"] for o in orders.values()),
    "total_rub": sum(o["total_rub"] for o in orders.values())
}

#     
print("  :")
for num, order in orders.items():
    print(f"\n{num}:")
    print(f"   : {order['output']} .")
    print(f"   : {order['norm_kg']} , {order['norm_rub']} .")
    print(f"   : {order['actual_kg']} , {order['actual_rub']} .")
    print(f"   : {order['waste_kg']:.1f} , {order['waste_rub']:.1f} .")
    print(f"   : {order['total_kg']:.1f} , {order['total_rub']:.1f} .")

print("\n :")
print(f"  : {totals['output']} .")
print(f"  : {totals['norm_kg']} , {totals['norm_rub']} .")
print(f"  : {totals['actual_kg']} , {totals['actual_rub']} .")
print(f"  : {totals['waste_kg']} , {totals['waste_rub']} .")
print(f"  : {totals['total_kg']:.1f} , {totals['total_rub']:.1f} .")
```
