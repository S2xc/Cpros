```python
# Исходные данные для продукта A и B
data = {
    "Продукт A": {
        "Материальные затраты": 1000,
        "Основная заработная плата": 1800,
        "Дополнительная заработная плата": 0.10,  # 10% от основной
        "Отчисления на соц. страхование": 0.30,  # 30% от ФОТ
        "Объем выпуска": 100
    },
    "Продукт B": {
        "Материальные затраты": 1500,
        "Основная заработная плата": 2000,
        "Дополнительная заработная плата": 0.08,  # 8% от основной
        "Отчисления на соц. страхование": 0.30,  # 30% от ФОТ
        "Объем выпуска": 200
    },
    "Накладные расходы": 1200
}

# Расчет дополнительной заработной платы
for product in ["Продукт A", "Продукт B"]:
    data[product]["Дополнительная заработная плата (руб)"] = data[product]["Основная заработная плата"] * data[product]["Дополнительная заработная плата"]

# Расчет ФОТ (основная + дополнительная заработная плата)
for product in ["Продукт A", "Продукт B"]:
    data[product]["ФОТ"] = data[product]["Основная заработная плата"] + data[product]["Дополнительная заработная плата (руб)"]

# Расчет отчислений на соц. страхование
for product in ["Продукт A", "Продукт B"]:
    data[product]["Отчисления на соц. страхование (руб)"] = data[product]["ФОТ"] * data[product]["Отчисления на соц. страхование"]

# Распределение накладных расходов пропорционально основной заработной плате
total_main_salary = data["Продукт A"]["Основная заработная плата"] + data["Продукт B"]["Основная заработная плата"]
data["Продукт A"]["Накладные расходы"] = (data["Продукт A"]["Основная заработная плата"] / total_main_salary) * data["Накладные расходы"]
data["Продукт B"]["Накладные расходы"] = (data["Продукт B"]["Основная заработная плата"] / total_main_salary) * data["Накладные расходы"]

# Расчет полной себестоимости для каждого продукта
for product in ["Продукт A", "Продукт B"]:
    data[product]["Полная себестоимость"] = (
        data[product]["Материальные затраты"] +
        data[product]["Основная заработная плата"] +
        data[product]["Дополнительная заработная плата (руб)"] +
        data[product]["Отчисления на соц. страхование (руб)"] +
        data[product]["Накладные расходы"]
    )

# Расчет себестоимости единицы продукции
for product in ["Продукт A", "Продукт B"]:
    data[product]["Себестоимость единицы"] = data[product]["Полная себестоимость"] / data[product]["Объем выпуска"]

# Вывод результатов
print("Результаты расчета для Продукта A:")
print(f"Материальные затраты: {data['Продукт A']['Материальные затраты']} руб.")
print(f"Основная заработная плата: {data['Продукт A']['Основная заработная плата']} руб.")
print(f"Дополнительная заработная плата: {data['Продукт A']['Дополнительная заработная плата (руб)']:.2f} руб.")
print(f"ФОТ: {data['Продукт A']['ФОТ']:.2f} руб.")
print(f"Отчисления на соц. страхование: {data['Продукт A']['Отчисления на соц. страхование (руб)']:.2f} руб.")
print(f"Накладные расходы: {data['Продукт A']['Накладные расходы']:.2f} руб.")
print(f"Полная себестоимость: {data['Продукт A']['Полная себестоимость']:.2f} руб.")
print(f"Себестоимость единицы: {data['Продукт A']['Себестоимость единицы']:.2f} руб./шт.")

print("\nРезультаты расчета для Продукта B:")
print(f"Материальные затраты: {data['Продукт B']['Материальные затраты']} руб.")
print(f"Основная заработная плата: {data['Продукт B']['Основная заработная плата']} руб.")
print(f"Дополнительная заработная плата: {data['Продукт B']['Дополнительная заработная плата (руб)']:.2f} руб.")
print(f"ФОТ: {data['Продукт B']['ФОТ']:.2f} руб.")
print(f"Отчисления на соц. страхование: {data['Продукт B']['Отчисления на соц. страхование (руб)']:.2f} руб.")
print(f"Накладные расходы: {data['Продукт B']['Накладные расходы']:.2f} руб.")
print(f"Полная себестоимость: {data['Продукт B']['Полная себестоимость']:.2f} руб.")
print(f"Себестоимость единицы: {data['Продукт B']['Себестоимость единицы']:.2f} руб./шт.")

# Расчет итоговых значений
total_material = data["Продукт A"]["Материальные затраты"] + data["Продукт B"]["Материальные затраты"]
total_main_salary = data["Продукт A"]["Основная заработная плата"] + data["Продукт B"]["Основная заработная плата"]
total_additional_salary = data["Продукт A"]["Дополнительная заработная плата (руб)"] + data["Продукт B"]["Дополнительная заработная плата (руб)"]
total_social_insurance = data["Продукт A"]["Отчисления на соц. страхование (руб)"] + data["Продукт B"]["Отчисления на соц. страхование (руб)"]
total_overhead = data["Накладные расходы"]
total_cost = data["Продукт A"]["Полная себестоимость"] + data["Продукт B"]["Полная себестоимость"]

print("\nИтоговые значения:")
print(f"Материальные затраты (Итого): {total_material} руб.")
print(f"Основная заработная плата (Итого): {total_main_salary} руб.")
print(f"Дополнительная заработная плата (Итого): {total_additional_salary:.2f} руб.")
print(f"Отчисления на соц. страхование (Итого): {total_social_insurance:.2f} руб.")
print(f"Накладные расходы (Итого): {total_overhead} руб.")
print(f"Общая полная себестоимость (Итого): {total_cost:.2f} руб.")
```

```python
# Исходные данные
data = {
    "Продукт A": {
        "Материальные затраты": 5000,
        "Основная заработная плата": 7800,
        "Дополнительная заработная плата": 0.10,  # 10% от основной
        "Отчисления на соц. страхование": 0.30,  # 30% от ФОТ
        "Коммерческие расходы": 0.02,  # 2% от производственной себестоимости
        "Объем выпуска": 100
    },
    "Продукт Б": {
        "Материальные затраты": 6500,
        "Основная заработная плата": 8000,
        "Дополнительная заработная плата": 0.08,  # 8% от основной
        "Отчисления на соц. страхование": 0.30,  # 30% от ФОТ
        "Коммерческие расходы": 0.05,  # 5% от производственной себестоимости
        "Объем выпуска": 200
    },
    "Общепроизводственные расходы": 10200,
    "Общехозяйственные расходы": 15000
}

# Расчет дополнительной заработной платы
for product in ["Продукт A", "Продукт Б"]:
    data[product]["Дополнительная заработная плата (руб)"] = data[product]["Основная заработная плата"] * data[product]["Дополнительная заработная плата"]

# Расчет ФОТ (основная + дополнительная заработная плата)
for product in ["Продукт A", "Продукт Б"]:
    data[product]["ФОТ"] = data[product]["Основная заработная плата"] + data[product]["Дополнительная заработная плата (руб)"]

# Расчет отчислений на соц. страхование
for product in ["Продукт A", "Продукт Б"]:
    data[product]["Отчисления на соц. страхование (руб)"] = data[product]["ФОТ"] * data[product]["Отчисления на соц. страхование"]

# Расчет прямых затрат (материалы + ФОТ + соц. отчисления)
for product in ["Продукт A", "Продукт Б"]:
    data[product]["Прямые затраты"] = (
        data[product]["Материальные затраты"] +
        data[product]["ФОТ"] +
        data[product]["Отчисления на соц. страхование (руб)"]
    )

# Распределение общепроизводственных расходов пропорционально прямым затратам
total_direct_costs = data["Продукт A"]["Прямые затраты"] + data["Продукт Б"]["Прямые затраты"]
data["Продукт A"]["Общепроизводственные расходы"] = (data["Продукт A"]["Прямые затраты"] / total_direct_costs) * data["Общепроизводственные расходы"]
data["Продукт Б"]["Общепроизводственные расходы"] = (data["Продукт Б"]["Прямые затраты"] / total_direct_costs) * data["Общепроизводственные расходы"]

# Распределение общехозяйственных расходов пропорционально прямым затратам
data["Продукт A"]["Общехозяйственные расходы"] = (data["Продукт A"]["Прямые затраты"] / total_direct_costs) * data["Общехозяйственные расходы"]
data["Продукт Б"]["Общехозяйственные расходы"] = (data["Продукт Б"]["Прямые затраты"] / total_direct_costs) * data["Общехозяйственные расходы"]

# Расчет производственной себестоимости
for product in ["Продукт A", "Продукт Б"]:
    data[product]["Производственная себестоимость"] = (
        data[product]["Прямые затраты"] +
        data[product]["Общепроизводственные расходы"] +
        data[product]["Общехозяйственные расходы"]
    )

# Расчет коммерческих расходов
for product in ["Продукт A", "Продукт Б"]:
    data[product]["Коммерческие расходы (руб)"] = data[product]["Производственная себестоимость"] * data[product]["Коммерческие расходы"]

# Расчет полной себестоимости
for product in ["Продукт A", "Продукт Б"]:
    data[product]["Полная себестоимость"] = (
        data[product]["Производственная себестоимость"] +
        data[product]["Коммерческие расходы (руб)"]
    )

# Расчет себестоимости единицы продукции
for product in ["Продукт A", "Продукт Б"]:
    data[product]["Себестоимость единицы"] = data[product]["Полная себестоимость"] / data[product]["Объем выпуска"]

# Вывод результатов
print("Результаты расчета для Продукта A:")
print(f"Материальные затраты: {data['Продукт A']['Материальные затраты']} руб.")
print(f"Основная заработная плата: {data['Продукт A']['Основная заработная плата']} руб.")
print(f"Дополнительная заработная плата: {data['Продукт A']['Дополнительная заработная плата (руб)']:.2f} руб.")
print(f"ФОТ: {data['Продукт A']['ФОТ']:.2f} руб.")
print(f"Отчисления на соц. страхование: {data['Продукт A']['Отчисления на соц. страхование (руб)']:.2f} руб.")
print(f"Прямые затраты: {data['Продукт A']['Прямые затраты']:.2f} руб.")
print(f"Общепроизводственные расходы: {data['Продукт A']['Общепроизводственные расходы']:.2f} руб.")
print(f"Общехозяйственные расходы: {data['Продукт A']['Общехозяйственные расходы']:.2f} руб.")
print(f"Производственная себестоимость: {data['Продукт A']['Производственная себестоимость']:.2f} руб.")
print(f"Коммерческие расходы: {data['Продукт A']['Коммерческие расходы (руб)']:.2f} руб.")
print(f"Полная себестоимость: {data['Продукт A']['Полная себестоимость']:.2f} руб.")
print(f"Себестоимость единицы: {data['Продукт A']['Себестоимость единицы']:.2f} руб./шт.")

print("\nРезультаты расчета для Продукта Б:")
print(f"Материальные затраты: {data['Продукт Б']['Материальные затраты']} руб.")
print(f"Основная заработная плата: {data['Продукт Б']['Основная заработная плата']} руб.")
print(f"Дополнительная заработная плата: {data['Продукт Б']['Дополнительная заработная плата (руб)']:.2f} руб.")
print(f"ФОТ: {data['Продукт Б']['ФОТ']:.2f} руб.")
print(f"Отчисления на соц. страхование: {data['Продукт Б']['Отчисления на соц. страхование (руб)']:.2f} руб.")
print(f"Прямые затраты: {data['Продукт Б']['Прямые затраты']:.2f} руб.")
print(f"Общепроизводственные расходы: {data['Продукт Б']['Общепроизводственные расходы']:.2f} руб.")
print(f"Общехозяйственные расходы: {data['Продукт Б']['Общехозяйственные расходы']:.2f} руб.")
print(f"Производственная себестоимость: {data['Продукт Б']['Производственная себестоимость']:.2f} руб.")
print(f"Коммерческие расходы: {data['Продукт Б']['Коммерческие расходы (руб)']:.2f} руб.")
print(f"Полная себестоимость: {data['Продукт Б']['Полная себестоимость']:.2f} руб.")
print(f"Себестоимость единицы: {data['Продукт Б']['Себестоимость единицы']:.2f} руб./шт.")

# Расчет итоговых значений
total_material = data["Продукт A"]["Материальные затраты"] + data["Продукт Б"]["Материальные затраты"]
total_main_salary = data["Продукт A"]["Основная заработная плата"] + data["Продукт Б"]["Основная заработная плата"]
total_additional_salary = data["Продукт A"]["Дополнительная заработная плата (руб)"] + data["Продукт Б"]["Дополнительная заработная плата (руб)"]
total_social_insurance = data["Продукт A"]["Отчисления на соц. страхование (руб)"] + data["Продукт Б"]["Отчисления на соц. страхование (руб)"]
total_direct = data["Продукт A"]["Прямые затраты"] + data["Продукт Б"]["Прямые затраты"]
total_overhead = data["Общепроизводственные расходы"] + data["Общехозяйственные расходы"]
total_cost = data["Продукт A"]["Полная себестоимость"] + data["Продукт Б"]["Полная себестоимость"]

print("\nИтоговые значения:")
print(f"Материальные затраты (Итого): {total_material} руб.")
print(f"Основная заработная плата (Итого): {total_main_salary} руб.")
print(f"Дополнительная заработная плата (Итого): {total_additional_salary:.2f} руб.")
print(f"Отчисления на соц. страхование (Итого): {total_social_insurance:.2f} руб.")
print(f"Прямые затраты (Итого): {total_direct:.2f} руб.")
print(f"Общепроизводственные и общехозяйственные расходы (Итого): {total_overhead} руб.")
print(f"Общая полная себестоимость (Итого): {total_cost:.2f} руб.")
```

График распределения затрат и расчеты представлены в коде выше.

```python
# Исходные данные
price_per_kg = 2000  # Цена стали за кг
waste_price = 800    # Цена отходов за кг
total_waste = 3000    # Общее количество отходов в кг

# Данные по заказам
orders = {
    "№1": {"output": 3000, "norm_rate": 4, "actual_rate": 4},
    "№2": {"output": 2000, "norm_rate": 20, "actual_rate": 17},
    "№3": {"output": 1000, "norm_rate": 2, "actual_rate": 3}
}

# 1. Расчет нормативного расхода
for order in orders.values():
    order["norm_kg"] = order["output"] * order["norm_rate"]
    order["norm_rub"] = order["norm_kg"] * price_per_kg

# 2. Расчет фактического расхода
for order in orders.values():
    order["actual_kg"] = order["output"] * order["actual_rate"]
    order["actual_rub"] = order["actual_kg"] * price_per_kg

# 3. Расчет общего нормативного расхода для распределения отходов
total_norm_kg = sum(o["norm_kg"] for o in orders.values())

# 4. Распределение отходов
for order in orders.values():
    order["waste_kg"] = (order["norm_kg"] / total_norm_kg) * total_waste
    order["waste_rub"] = order["waste_kg"] * waste_price

# 5. Расчет итоговых затрат
for order in orders.values():
    order["total_kg"] = order["actual_kg"] - order["waste_kg"]
    order["total_rub"] = order["actual_rub"] - order["waste_rub"]

# 6. Расчет общих итогов
totals = {
    "output": sum(o["output"] for o in orders.values()),
    "norm_kg": sum(o["norm_kg"] for o in orders.values()),
    "norm_rub": sum(o["norm_rub"] for o in orders.values()),
    "actual_kg": sum(o["actual_kg"] for o in orders.values()),
    "actual_rub": sum(o["actual_rub"] for o in orders.values()),
    "waste_kg": total_waste,
    "waste_rub": total_waste * waste_price,
    "total_kg": sum(o["total_kg"] for o in orders.values()),
    "total_rub": sum(o["total_rub"] for o in orders.values())
}

# Вывод данных в удобном формате
print("Данные по заказам:")
for num, order in orders.items():
    print(f"\n{num}:")
    print(f"  Выход продукции: {order['output']} шт.")
    print(f"  Нормативный расход: {order['norm_kg']} кг, {order['norm_rub']} руб.")
    print(f"  Фактический расход: {order['actual_kg']} кг, {order['actual_rub']} руб.")
    print(f"  Возврат отходов: {order['waste_kg']:.1f} кг, {order['waste_rub']:.1f} руб.")
    print(f"  Итого затрат: {order['total_kg']:.1f} кг, {order['total_rub']:.1f} руб.")

print("\nИтоговые значения:")
print(f"Общий выход продукции: {totals['output']} шт.")
print(f"Суммарный нормативный расход: {totals['norm_kg']} кг, {totals['norm_rub']} руб.")
print(f"Суммарный фактический расход: {totals['actual_kg']} кг, {totals['actual_rub']} руб.")
print(f"Общий возврат отходов: {totals['waste_kg']} кг, {totals['waste_rub']} руб.")
print(f"Общие итоговые затраты: {totals['total_kg']:.1f} кг, {totals['total_rub']:.1f} руб.")
```

```python
# Исходные данные
data = {
    "Изделие A": {
        "Выручка": 5000000,
        "Зарплата": 900000,
        "Материалы": 1400000,
        "Переменные_накладные": 600000
    },
    "Изделие Б": {
        "Выручка": 1500000,
        "Зарплата": 600000,
        "Материалы": 300000,
        "Переменные_накладные": 200000
    },
    "Постоянные_накладные": 1800000
}

# Расчет суммы покрытия (маржинального дохода)
for product in ["Изделие A", "Изделие Б"]:
    # Сумма покрытия = Выручка - Переменные затраты
    variable_costs = (data[product]["Зарплата"] + 
                     data[product]["Материалы"] + 
                     data[product]["Переменные_накладные"])
    data[product]["Сумма_покрытия"] = data[product]["Выручка"] - variable_costs
    data[product]["Коэф_покрытия"] = data[product]["Сумма_покрытия"] / data[product]["Выручка"]

# Расчет результата работы предприятия
total_contribution = data["Изделие A"]["Сумма_покрытия"] + data["Изделие Б"]["Сумма_покрытия"]
profit = total_contribution - data["Постоянные_накладные"]

# Формирование таблицы результатов
print("| Показатель               | Изделие A     | Изделие Б     |")
print("|--------------------------|---------------|---------------|")
print(f"| Выручка                 | {data['Изделие A']['Выручка']:,} | {data['Изделие Б']['Выручка']:,} |")
print(f"| Переменные затраты      | {data['Изделие A']['Зарплата'] + data['Изделие A']['Материалы'] + data['Изделие A']['Переменные_накладные']:,} | {data['Изделие Б']['Зарплата'] + data['Изделие Б']['Материалы'] + data['Изделие Б']['Переменные_накладные']:,} |")
print(f"| Сумма покрытия          | {data['Изделие A']['Сумма_покрытия']:,} | {data['Изделие Б']['Сумма_покрытия']:,} |")
print(f"| Коэффициент покрытия    | {data['Изделие A']['Коэф_покрытия']:.2f}        | {data['Изделие Б']['Коэф_покрытия']:.2f}        |")
print("\nОбщие постоянные накладные расходы:", f"{data['Постоянные_накладные']:,}")
print("Прибыль предприятия:", f"{profit:,}")
```
