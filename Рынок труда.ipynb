```python
from sympy import symbols, Eq, solve

# Объявляем переменные
w = symbols('w')  # ставка оплаты труда
DL = 102 - 2 * w  # функция спроса
SL = 6 + w  # функция предложения

# Равновесие: спрос = предложение
equation = Eq(DL, SL)

# Решаем уравнение
w_solution = solve(equation, w)[0]

# Находим равновесное количество занятых работников
L = DL.subs(w, w_solution)

print(f"Равновесная ставка оплаты труда: {w_solution}")
print(f"Равновесное количество занятых работников: {L} тыс. человек")
```

### Итог:

Равновесная ставка оплаты труда составляет 32, а равновесное количество занятых работников — 38 тысяч человек.

```python
# Объявляем новые переменные
w_min = 30  # минимальная ставка заработной платы
DL_new = 70 - w  # новая функция спроса
SL_new = 4 * w - 30  # новая функция предложения

# Находим равновесие без минимальной ставки
equation_no_min = Eq(DL_new, SL_new)
w_eq_no_min = solve(equation_no_min, w)[0]
L_eq_no_min = DL_new.subs(w, w_eq_no_min)

# Рассчитываем состояние рынка при минимальной ставке
L_demand_min = DL_new.subs(w, w_min)  # спрос при минимальной ставке
L_supply_min = SL_new.subs(w, w_min)  # предложение при минимальной ставке

w_eq_no_min, L_eq_no_min, L_demand_min, L_supply_min
```

Без установления минимальной ставки заработной платы равновесная ставка составила бы 20 ДЕ в час, а равновесное количество занятых работников — 50 тыс. человек.

При минимальной ставке заработной платы в 30 ДЕ в час:
	•	Спрос на труд составляет 40 тыс. человек.
	•	Предложение труда составляет 90 тыс. человек.

Это приводит к избытку предложения труда на рынке, так как количество людей, желающих работать по минимальной ставке, значительно превышает количество вакансий, что может вызвать безработицу среди части работников. ￼

```python
import numpy as np

# Данные
Y = 4_000_000_000  # фактический ВВП в ДЕ
L = 350_000_000    # численность рабочей силы
U = 30_000_000     # безработные
B = 2              # коэффициент Оукена

# Подсчёт уровней безработицы
u = U / L  # фактический уровень безработицы

# Естественная безработица (в млн)
natural_unemployed = 3 + 5 + 4  # обучение, переориентация, уход
u_star = natural_unemployed / L

# Циклическая безработица
u_c = u - u_star

# Потенциальный ВВП по закону Оукена
Y_star = Y / (1 + B * u_c)

# Потери экономики
loss = Y_star - Y

# Вывод
print(f"Фактический уровень безработицы: {u:.4f} ({u*100:.2f}%)")
print(f"Естественный уровень безработицы: {u_star:.4f} ({u_star*100:.2f}%)")
print(f"Циклический уровень безработицы: {u_c:.4f} ({u_c*100:.2f}%)")
print(f"Потенциальный ВВП: {Y_star:.2f} ДЕ")
print(f"Потери экономики: {loss:.2f} ДЕ")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные: [доля населения, средний доход в группе]
groups = [
    (16.0, 5000),
    (12.5, 6000),   # среднее между 5000.1 и 7000
    (12.0, 8000),
    (15.0, 10500),
    (11.0, 13500),
    (12.3, 17500),
    (7.5, 22500),
    (5.0, 27500),
    (3.0, 32500)
]

# Преобразуем в массивы
population_shares = np.array([g[0] for g in groups]) / 100  # в долях
average_incomes = np.array([g[1] for g in groups])

# Вычисляем общий доход в каждой группе
group_incomes = population_shares * average_incomes

# Сортировка по доходу (уже отсортированы, но на всякий случай)
sorted_indices = np.argsort(average_incomes)
population_shares = population_shares[sorted_indices]
group_incomes = group_incomes[sorted_indices]

# Кумулятивные доли населения и доходов
cum_population = np.cumsum(population_shares)
cum_income = np.cumsum(group_incomes) / np.sum(group_incomes)

# Добавляем точку (0,0) для корректной кривой Лоренца
cum_population = np.insert(cum_population, 0, 0)
cum_income = np.insert(cum_income, 0, 0)

# Построение кривой Лоренца
plt.figure(figsize=(6, 6))
plt.plot(cum_population, cum_income, drawstyle='steps-post', label='Кривая Лоренца')
plt.plot([0, 1], [0, 1], '--', color='gray', label='Линия абсолютного равенства')
plt.title('Кривая Лоренца')
plt.xlabel('Кумулятивная доля населения')
plt.ylabel('Кумулятивная доля дохода')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()

# Расчёт коэффициента Джини (метод трапеций)
gini = 1 - 2 * np.trapz(cum_income, cum_population)
print(f"Коэффициент Джини: {gini:.4f}")
```
