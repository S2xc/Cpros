```python
import matplotlib.pyplot as plt

#   
prices = [0, 1, 2, 3, 4]
firm1_demand = [6, 5, 4, 3, 0]
firm2_demand = [6, 4, 3, 2, 0]
firm3_demand = [5, 4, 2, 1, 0]
market_demand = [17, 13, 9, 6, 0]

#  
plt.figure(figsize=(10, 6))

plt.plot(prices, firm1_demand, label=' 1')
plt.plot(prices, firm2_demand, label=' 2')
plt.plot(prices, firm3_demand, label=' 3')
plt.plot(prices, market_demand, label=' ', linestyle='--', linewidth=2)

plt.xlabel(', ')
plt.ylabel(' ,   ')
plt.title('    ')
plt.legend()
plt.grid(True)
plt.show()
```
