Чтобы определить равновесный уровень цены и равновесный объём продажи товара на рынке, нам нужно приравнять уравнения спроса и предложения.

### Уравнения спроса и предложения

- **Спрос (QD)**: $$QD = 30 - P$$
- **Предложение (QS)**: $$QS = 4P - 20$$

### Равновесие

Равновесие наступает, когда $$QD = QS$$. Приравнивая уравнения:

$$30 - P = 4P - 20$$

Решая для $$P$$:

$$30 + 20 = 5P$$
$$50 = 5P$$
$$P = 10$$

Подставляя $$P = 10$$ в любое из уравнений, чтобы найти равновесный объём ($$Q$$):

$$Q = 30 - 10 = 20$$

Итак, **равновесная цена** равна **10 ДЕ**, а **равновесный объём** — **20 единиц**.

### Административное регулирование цены

#### а) Цена 8 ДЕ

Если административно установить цену **8 ДЕ**, то:

- **Спрос**: $$QD = 30 - 8 = 22$$
- **Предложение**: $$QS = 4 \times 8 - 20 = 32 - 20 = 12$$

В этом случае на рынке возникнет **дефицит** товара, так как спрос (22 единицы) превышает предложение (12 единиц). Производители будут неохотно продавать по заниженной цене, а потребители будут готовы купить больше, чем производители готовы предложить.

#### б) Цена 12 ДЕ

Если административно установить цену **12 ДЕ**, то:

- **Спрос**: $$QD = 30 - 12 = 18$$
- **Предложение**: $$QS = 4 \times 12 - 20 = 48 - 20 = 28$$

В этом случае на рынке возникнет **избыток** товара, так как предложение (28 единиц) превышает спрос (18 единиц). Производители будут предлагать больше, чем потребители готовы купить, что приведет к снижению цены для стимулирования продаж.

### Вывод

Административное регулирование цены может привести к дисбалансу на рынке. Если цена установлена ниже равновесной (8 ДЕ), возникает дефицит; если выше равновесной (12 ДЕ), возникает избыток. Рыночное равновесие обеспечивает баланс между спросом и предложением, и любое административное вмешательство может нарушить этот баланс.

Чтобы проанализировать ситуацию на рынке при административной установке цены в 500 ДЕ, нам нужно сначала определить равновесную цену и объём, а затем сравнить их с ситуацией при установленной цене.

### Уравнения спроса и предложения

- **Спрос (QD)**: $$QD = 8000 - 12P$$
- **Предложение (QS)**: $$QS = 4P - 750$$

### Равновесие

Равновесие наступает, когда $$QD = QS$$. Приравнивая уравнения:

$$8000 - 12P = 4P - 750$$

Решая для $$P$$:

$$8000 + 750 = 16P$$
$$8750 = 16P$$
$$P = \frac{8750}{16} \approx 546,88$$

Подставляя $$P \approx 546,88$$ в любое из уравнений, чтобы найти равновесный объём ($$Q$$):

$$Q = 8000 - 12 \times 546,88 \approx 8000 - 6566,56 \approx 1433,44$$

Итак, **равновесная цена** примерно равна **546,88 ДЕ**, а **равновесный объём** — примерно **1433,44 единиц**.

### Административное регулирование цены

Если административно установить цену **500 ДЕ**, то:

- **Спрос**: $$QD = 8000 - 12 \times 500 = 8000 - 6000 = 2000$$
- **Предложение**: $$QS = 4 \times 500 - 750 = 2000 - 750 = 1250$$

В этом случае на рынке возникнет **дефицит** товара, так как спрос (2000 единиц) превышает предложение (1250 единиц). Производители будут неохотно продавать по заниженной цене, а потребители будут готовы купить больше, чем производители готовы предложить.

### Вывод

Административная установка цены в 500 ДЕ приведет к дефициту на рынке, поскольку спрос превышает предложение. Это может стимулировать рост цен в долгосрочной перспективе, если не будут приняты меры по увеличению производства или снижению спроса.

Если государство вводит 10-процентный налог с продаж, уплачиваемый продавцом, это означает, что продавец получит меньшую цену за единицу товара, чем та, которую платит покупатель. Чтобы проанализировать последствия такого налога, нам нужно понять, как он повлияет на кривые спроса и предложения.

### Исходные уравнения

- **Спрос (QD)**: $$QD = 100 - P$$
- **Предложение (QS)**: $$QS = 2P - 50$$

### Введение налога

Налог с продаж в размере 10% означает, что продавец получит только 90% от цены, которую платит покупатель. Если обозначить цену, которую платит покупатель, как $$P_b$$, а цену, которую получает продавец, как $$P_s$$, то:

$$P_s = P_b \times 0.9$$

или

$$P_b = \frac{P_s}{0.9}$$

### Налог и предложение

Поскольку налог уплачивается продавцом, он увеличивает издержки продавца. Это эквивалентно сдвигу кривой предложения вверх. Новое уравнение предложения будет учитывать налог:

$$QS = 2P_s - 50$$

Но поскольку $$P_s = 0.9P_b$$, то:

$$QS = 2 \times 0.9P_b - 50 = 1.8P_b - 50$$

### Налог и спрос

Кривая спроса остаётся прежней, так как налог не влияет напрямую на готовность покупателя платить:

$$QD = 100 - P_b$$

### Равновесие до и после налога

#### До налога

Равновесие наступает, когда $$QD = QS$$:

$$100 - P = 2P - 50$$

Решая для $$P$$:

$$100 + 50 = 3P$$
$$150 = 3P$$
$$P = 50$$

Подставляя $$P = 50$$ в любое из уравнений, чтобы найти равновесный объём ($$Q$$):

$$Q = 100 - 50 = 50$$

#### После налога

Новое уравнение равновесия с учётом налога:

$$100 - P_b = 1.8P_b - 50$$

Решая для $$P_b$$:

$$100 + 50 = 2.8P_b$$
$$150 = 2.8P_b$$
$$P_b = \frac{150}{2.8} \approx 53.57$$

Цена, получаемая продавцом ($$P_s$$):

$$P_s = 0.9 \times P_b \approx 0.9 \times 53.57 \approx 48.21$$

Подставляя $$P_b \approx 53.57$$ в уравнение спроса, чтобы найти новый равновесный объём ($$Q$$):

$$Q = 100 - 53.57 \approx 46.43$$

### Последствия

1. **Цена для покупателя** увеличивается с 50 ДЕ до примерно 53,57 ДЕ.
2. **Цена для продавца** уменьшается с 50 ДЕ до примерно 48,21 ДЕ.
3. **Равновесный объём** уменьшается с 50 единиц до примерно 46,43 единиц.

Введение налога с продаж приведет к увеличению цены для покупателя, снижению цены для продавца и уменьшению равновесного объёма продаж. Это может привести к снижению производства и продаж из-за повышения издержек для продавцов.

Чтобы найти новую равновесную цену ($$P_{E}^*$$) и новый равновесный объём ($$Q_{E}^*$$) после введения налога в размере 2 ДЕ за единицу блага, нам нужно скорректировать уравнение предложения, так как налог уплачивается продавцами.

### Исходные уравнения

- **Спрос (QD)**: $$QD = 70 - 2P$$
- **Предложение (QS)**: $$QS = 2P - 50$$

### Введение налога

Налог в размере 2 ДЕ за единицу блага означает, что продавцы получат на 2 ДЕ меньше, чем цена, которую платят покупатели. Если обозначить цену, которую платит покупатель, как $$P_b$$, а цену, которую получает продавец, как $$P_s$$, то:

$$P_s = P_b - 2$$

### Корректировка уравнения предложения

Новое уравнение предложения будет учитывать налог:

$$QS = 2P_s - 50 = 2(P_b - 2) - 50 = 2P_b - 4 - 50 = 2P_b - 54$$

### Равновесие после введения налога

Равновесие наступает, когда $$QD = QS$$:

$$70 - 2P_b = 2P_b - 54$$

Решая для $$P_b$$:

$$70 + 54 = 4P_b$$
$$124 = 4P_b$$
$$P_b = \frac{124}{4} = 31$$

Подставляя $$P_b = 31$$ в уравнение спроса, чтобы найти новый равновесный объём ($$Q_{E}^*$$):

$$Q_{E}^* = 70 - 2 \times 31 = 70 - 62 = 8$$

### Цена для продавца

Цена, получаемая продавцом ($$P_s$$):

$$P_s = P_b - 2 = 31 - 2 = 29$$

### Сумма налога

Сумма налога ($$T$$), поступающая в бюджет, рассчитывается как:

$$T = t \times Q_{E}^* = 2 \times 8 = 16$$

### Вывод

- **Новая равновесная цена ($$P_{E}^*$$)**: 31 ДЕ.
- **Новый равновесный объём ($$Q_{E}^*$$)**: 8 единиц.
- **Сумма налога ($$T$$)**: 16 ДЕ.

Введение налога в размере 2 ДЕ за единицу блага увеличивает цену для покупателя, снижает цену для продавца и уменьшает равновесный объём продаж.

Чтобы проанализировать изменения в равновесной цене и равновесном объёме после введения субсидии производителю, нам нужно скорректировать уравнение предложения, так как субсидия снижает издержки производителя.

### Исходные уравнения

- **Спрос (QD)**: $$QD = 1200 - 5P$$
- **Предложение (QS)**: $$QS = 500 + 5P$$

### Введение субсидии

Субсидия в размере 10 ДЕ на одну единицу продукции означает, что производитель получит на 10 ДЕ больше, чем цена, которую он получал бы без субсидии. Если обозначить цену, которую платит покупатель, как $$P_b$$, а цену, которую получает производитель, как $$P_s$$, то:

$$P_s = P_b + 10$$

### Корректировка уравнения предложения

Новое уравнение предложения будет учитывать субсидию:

$$QS = 500 + 5(P_b - 10) = 500 + 5P_b - 50 = 450 + 5P_b$$

### Равновесие после введения субсидии

Равновесие наступает, когда $$QD = QS$$:

$$1200 - 5P_b = 450 + 5P_b$$

Решая для $$P_b$$:

$$1200 - 450 = 10P_b$$
$$750 = 10P_b$$
$$P_b = \frac{750}{10} = 75$$

Подставляя $$P_b = 75$$ в уравнение спроса, чтобы найти новый равновесный объём ($$Q_{E}^*$$):

$$Q_{E}^* = 1200 - 5 \times 75 = 1200 - 375 = 825$$

### Цена реализации продукции для производителя

Цена, получаемая производителем ($$P_s$$):

$$P_s = P_b + 10 = 75 + 10 = 85$$

### Размер расходов на субсидию

Размер расходов на субсидию ($$H$$) рассчитывается как:

$$H = h \times Q_{E}^* = 10 \times 825 = 8250$$

### Вывод

- **Новая равновесная цена ($$P_{E}^*$$)**: 75 ДЕ.
- **Новый равновесный объём ($$Q_{E}^*$$)**: 825 единиц.
- **Цена реализации продукции для производителя**: 85 ДЕ.
- **Размер расходов на субсидию ($$H$$)**: 8250 ДЕ.

Введение субсидии снижает цену для покупателя, увеличивает цену для производителя и увеличивает равновесный объём продаж.

Чтобы рассчитать коэффициент прямой эластичности спроса по цене, мы используем формулу:

$$ E_{P}^{D} = \frac{\text{Процентное изменение спроса}}{\text{Процентное изменение цены}} $$

В данном случае:

- **Процентное изменение спроса**: -3% (спрос снизился на 3%).
- **Процентное изменение цены**: 1,5% (цена увеличилась на 1,5%).

Подставляя эти значения в формулу:

$$ E_{P}^{D} = \frac{-3\%}{1,5\%} = -2 $$

Поскольку обычно знак минус опускается при оценке эластичности, коэффициент прямой эластичности спроса по цене равен:

$$ E_{P}^{D} = 2 $$

Это означает, что спрос на товар **эластичен** по цене, так как процентное изменение спроса больше процентного изменения цены.

Чтобы определить коэффициент эластичности спроса по доходу, нам нужно использовать формулу:

$$ E_{Y}^{D} = \frac{\text{Процентное изменение спроса}}{\text{Процентное изменение дохода}} $$

### Данные

- **Исходный доход**: 40 тыс. руб.
- **Исходный объём спроса**: 10 кг
- **Новый доход**: 54 тыс. руб.
- **Новый объём спроса**: 16 кг
- **Цена товара**: остаётся прежней (120 руб. за кг)

### Расчёт процентного изменения спроса и дохода

- **Процентное изменение спроса**:
  $$
  \Delta Q = 16 - 10 = 6 \text{ кг}
  $$
  $$
  \text{Процентное изменение спроса} = \frac{\Delta Q}{Q_{\text{исход}}} \times 100 = \frac{6}{10} \times 100 = 60\%
  $$

- **Процентное изменение дохода**:
  $$
  \Delta Y = 54 - 40 = 14 \text{ тыс. руб.}
  $$
  $$
  \text{Процентное изменение дохода} = \frac{\Delta Y}{Y_{\text{исход}}} \times 100 = \frac{14}{40} \times 100 = 35\%
  $$

### Коэффициент эластичности спроса по доходу

$$ E_{Y}^{D} = \frac{60\%}{35\%} \approx 1,71 $$

### Тип товара

Поскольку коэффициент эластичности спроса по доходу больше 1, товар С можно классифицировать как **нормальный товар** с **эластичным спросом по доходу**. Это означает, что при увеличении дохода потребителя спрос на этот товар увеличивается более чем пропорционально.

Товары с таким типом эластичности часто являются предметами роскоши или имеют высокую чувствительность к изменениям дохода потребителей. Однако в данном случае, учитывая, что цена товара остаётся прежней, а спрос увеличивается с ростом дохода, это может указывать на то, что товар не является строго необходимым, но его потребление растёт с увеличением благосостояния потребителя.

Чтобы определить коэффициент перекрестной эластичности спроса на товар А по цене товара B, нам нужно использовать формулу:

$$ E_{AB}^{D} = \frac{\Delta Q_{A} / Q_{A}}{\Delta P_{B} / P_{B}} $$

Однако, поскольку у нас есть функция спроса, мы можем использовать производную для нахождения коэффициента перекрестной эластичности:

$$ Q_{DA} = -2P_{A} - 0,3P_{B} + 810 $$

### Производная по цене товара B

$$ \frac{\partial Q_{DA}}{\partial P_{B}} = -0,3 $$

### Рассчитываем исходный объём спроса на товар А

Подставляем данные значения $$ P_{A} = 300 $$ руб./ед. и $$ P_{B} = 200 $$ руб./ед. в функцию спроса:

$$ Q_{DA} = -2 \times 300 - 0,3 \times 200 + 810 = -600 - 60 + 810 = 150 $$

### Коэффициент перекрестной эластичности

Теперь мы можем рассчитать коэффициент перекрестной эластичности:

$$ E_{AB}^{D} = \frac{\frac{\partial Q_{DA}}{\partial P_{B}}}{\frac{Q_{DA}}{P_{B}}} = \frac{-0,3}{\frac{150}{200}} = \frac{-0,3}{0,75} = -0,4 $$

### Вывод о типе связи между товарами

Поскольку коэффициент перекрестной эластичности отрицательный ($$ E_{AB}^{D} < 0 $$), это означает, что товары А и B являются **взаимодополняющими**. Это значит, что при увеличении цены товара B спрос на товар А уменьшается, что характерно для товаров, которые потребляются вместе или являются частью одного набора. Примерами таких товаров могут быть автомобили и бензин или ноутбуки и комплектующие к ним.

### Чистое общественное благо

Для чистого общественного блага коллективный спрос рассчитывается как максимальная сумма, которую все потребители вместе готовы заплатить за единицу блага. Поскольку в таблице нет информации о готовности платить, мы можем предположить, что коллективный спрос будет равен сумме индивидуальных спросов, но это не совсем точно для общественных благ, где каждый потребитель может потреблять благо одновременно без уменьшения его доступности для других.

Однако, если мы рассматриваем готовность платить как сумму индивидуальных спросов (что не является точным подходом для общественных благ), то коллективный спрос можно представить как сумму индивидуальных спросов при каждом уровне цены. Но для общественных благ обычно используется другой подход, учитывающий, что все потребители могут потреблять благо одновременно.
